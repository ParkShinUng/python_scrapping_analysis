```python
def clean_text(text): # 문자열에 있는 특수문자 제거 함수
    import re 
    text = text.replace("\n", "")
    cleaned_text = re.sub('[\{\}\[\]\/?.,;:|\)*~`!^\-_+<>@\#$%&\\\=\(\'\"]', '', text)
    cleaned_text = cleaned_text.replace(' ', '')
    
    return cleaned_text
```

### 팟빵 다운로드 1단계
* 팟빵 id와 제목을 함수의 인자로 전달
* 저장되는 mp3 파일명을 에피소드의 타이틀로 지정
* 모든 에피소드를 다운로드 받는다.
 - 에피소드를 요청할 때 offset 값의 최대값이 항상 변하므로
 - itertools의 count() 함수를 사용해 무한루프를 수행하도록 한다.
 - 무한루프를 빠져 나올수 있는 조건식을 반드시 주어야 한다.
   - json_data['data'] 길이가 0이면 무한루프를 탈출한다.
 - 사람이 직접 다운로드 받는 것 처럼 보이게 하기 위해서 sleep time 설정
   - time 이라는 내부모듈의 sleep이라는 함수 사용

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import json
import os

def podbbang(chNum, chTitle):

    url = 'http://www.podbbang.com/_m_api/podcasts/'+chNum+'/episodes?offset=0&sort=pubdate:desc&limit=8&with=summary&cache=0'

    res = requests.get(url)

    if res.ok:
        soup = BeautifulSoup(res.text, 'lxml')
        p_tag = soup.select('p')[0].text # str 타입의 데이터를 json(dict타입)으로 변환
        json_data = json.loads(p_tag)

        episode_length = len(json_data['data'])
        episode_list = json_data['data']

        for episode in episode_list:
            title = episode['title'] # 에피소드의 제목
            mp3_url = episode['enclosure']['url'] # 에피소드 다운로드 url

            if mp3_url: # url 값이 있으면 download
                req_header = {
                    'referer':'http://www.podbbang.com/ch/'+chNum, # 요청 시작페이지 전달
                    'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.111 Safari/537.36'
                }

                dir_name = os.path.join('mp3', chTitle) # mp3 여바라
                if not os.path.isdir(dir_name): # 해당 디렉토리 없으면 생성
                    os.makedirs(dir_name)

                file_name = os.path.basename(mp3_url) # 해당 url에서 파일명만 추출
                file_name = dir_name + '/' + file_name # 생성된 디렉토리명과 파일명 join

                res = requests.get(mp3_url, headers=req_header)
                if res.ok:
                    mp3_bin = res.content # response 객체에서 binary 데이터 추출
                    with open(file_name, 'wb') as file: # binary 데이터를 local 파일로 저장
                        file.write(mp3_bin)



podbbang('16898', '정영진 최욱의 매불쇼')
```

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import json
import os
from itertools import count
from time import sleep

def podbbang_all_download(pod_id, title):
    url = f'http://www.podbbang.com/_m_api/podcasts/{pod_id}/episodes'
    
    for offset in count(0):
        req_param = {
            "offset":offset,
            "sort":"pubdate:desc",
            "episode_id":0,
            "limit":8,
            "with":"summary",
            "cache":0
        }
        res = requests.get(url, params=req_param)

        if res.ok:
            soup = BeautifulSoup(res.text, 'lxml')
            p_tag = soup.select('p')[0].text # str 타입의 데이터를 json(dict타입)으로 변환
            json_data = json.loads(p_tag)

            episode_length = len(json_data['data'])
            print('offset : {} data의 길이 : {}'.format(offset, episode_length))
            if episode_length == 0: # 무한루프 탈출 조건을 설정
                return None
            episode_list = json_data['data']

            for episode in episode_list:
                episode_title = episode['title'] # 에피소드의 제목
                mp3_url = episode['enclosure']['url'] # 에피소드 다운로드 url

                if mp3_url: # url 값이 있으면 download
                    req_header = {
                        'referer':f'http://www.podbbang.com/ch/{pod_id}', # 요청 시작페이지 전달
                        'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.111 Safari/537.36'
                    }

                    dir_name = os.path.join('mp3', title) # mp3 + 인자 title
                    
                    if not os.path.isdir(dir_name): # 해당 디렉토리 없으면 생성
                        os.makedirs(dir_name)

                    episode_title = clean_text(episode_title)
                    file_name = f'{episode_title}.mp3'
                    file_name = dir_name + '/' + file_name # 생성된 디렉토리명과 파일명 join

                    res = requests.get(mp3_url, headers=req_header)
                    
                    if res.ok:
                        mp3_bin = res.content # response 객체에서 binary 데이터 추출
                        with open(file_name, 'wb') as file: # binary 데이터를 local 파일로 저장
                            file.write(mp3_bin)
                            
                        # 0.5초간 프로세스를 중지함, 기계가 아니라 사람처럼 보이게 하기 위해
                        sleep(0.5)

podbbang_all_download('16255', '여바라')
```
