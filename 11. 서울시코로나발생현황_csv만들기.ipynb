### 서울시 코로나 19 발생동향 분석
* https://www.seoul.go.kr/coronaV/coronaStatus.do
* pandas의 read_html() 함수를 사용하여 table 엘리먼트를 추출
* csv 파일로 저장하기

```python
import pandas as pd

url = 'https://www.seoul.go.kr/coronaV/coronaStatus.do'

table_df = pd.read_html(url)
print(len(table_df))
```

```python
print(table_df[3].shape)
data_df = table_df[3]
```

```python
# 마지막의 확진일 값을 파일명에 사용하기 위해 .을 _로 변경하기

last_day = data_df.loc[0, '확진일']
last_day = last_day.replace('.', '_')
print(last_day)

file_name = f'data/seoul_covid_{last_day}.csv'
file_name
```

```python
# csv 파일로 저장하기
#data_df.to_csv(file_name, index=False)
data_df.to_csv(file_name, index=False)
```
